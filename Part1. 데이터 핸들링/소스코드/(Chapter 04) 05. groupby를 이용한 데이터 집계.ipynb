```python
df = pd.read_csv("온라인_판매기록.csv", engine = "python")
```

```python
df.groupby(['쇼핑몰 유형'])['수량'].mean()
# 쇼핑몰 유형에 따른 수량의 평균.
#Name이 수량이라는 점. series라는 점. 
```

```python
df.groupby(['쇼핑몰 유형'], as_index = False)['수량'].mean()
```

```python
df.groupby(['쇼핑몰 유형'])['수량', '판매금액'].agg(['mean', 'max'])
# 동시에 두개 함수 사용하기 위해서는 agg 함수 사용. 
```

```python
def my_func(value):
    return max(value) - min(value)

df.groupby(['제품', '쇼핑몰 유형'])['수량', '판매금액'].agg(['mean', my_func])
# my func의 인자로는 수량과 판매금액이 들어가게 된다. 
```
