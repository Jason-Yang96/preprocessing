
#### 배열 만들기
##### Numpy 자료형은 ndarray로 효율적인 배열 연산을 하기 위해 개발되었음. 리스트와 ndarray는 유연성과 효율성을 기준으로 비교할 수 있음. 


```python
# 리스트에서 배열 만들기
import numpy as np
a = np.array([1, 4, 2, 5, 3])
b = np.array([3.14, 4, 2, 3])
c = np.array([1, 2, 3, 4], dtype = float)
print(a, b, c)
# dtype이 float인 경우는 뒤에 소수점이 붙어서 출력된다. 
```

```python
# 함수를 이용하여 배열 만들기
a = np.zeros( (10,10), dtype = int) # 0으로 채운 배열 생성 - 특정 숫자 대신에 shape을 반환하여 행과 열을 구성하여도 된다. 
b = np.ones((3, 5), dtype = float) #1로 채운 배열 생성
c = np.full((3,5), 3.14) # 3.14로 채운 배열 만들기 
d = np.arange(0, 20, 2) # 0에서 시작해 2씩 더해 20까지 채우는 배열 생성 - 끝자리 포함하지 않음. 
e = np.linspace(0, 1, 5) # 0과 1 사이에 일정한 간격을 가진 다섯 개의 값으로 채운 배열 만들기 - 끝자리 포함하기 됨. 특이한 캐이스
f = np.random.random((3,3)) # 3*3 크기의 난수 배열 생성 - 값은 0부터 1값으로 채워진다. 
g = np.random.normal(0, 1, (3, 3)) # 평균 0, 표준 편차 1의 정규 분포를 따르는 3*3 난수 배열 - 여기서는 표준정규분포가 된다. 
h = np.random.randint(0, 10, (3, 3)) # [0, 10) 구간의 임의로 정수로 채운 3*3 배열 만들기 - 구간 정수로 이루어진 array 반환하게 된다. 
i = np.eye(3) # 크기 3의 단위 행렬 만들기 - 주대각선의 값이 모두 1이고 나머지는 0인 행렬을 의미한다. 

print("np.zeros((10,10), dtype = int):\n", a)
print("\n np.ones((3, 5), dtype = float):\n", b)
print("\n np.full((3,5), 3.14):\n", c)
print("\n np.arange(0, 20, 2):\n", d)
print("\n np.linspace(0, 1, 5):\n", e)
print("\n np.random.random((3,3)):\n", f)
print("\n np.random.normal(0, 1, (3, 3)):\n", g)
print("\n np.random.randint(0, 10, (3, 3)):\n", h)
print("\n np.eye(3):\n", i)
```

#### 인덱싱과 슬라이싱: 기본적으로 리스트 자료형과 완전히 동일하다. 부울 리스트도 인덱스로 활용할 수 있다. True인 요소와 대응되는 요소만 가져온다. 여러 인덱스를 리스트 형대로 받아서 값을 찾을 수도 있다.

```python
# 기본적인 인덱싱과 슬라이싱
x1 = np.array([1, 3, 5, 7, 9])

print("x1[0]:", x1[0]) # 양수 인덱스 (맨 앞 요소)
print("x1[3]:", x1[3]) # 세 번째 요소 
print("x1[-1]:", x1[-1]) # 음수 인덱스 (맨 뒤 요소)
print("x1[-2]:", x1[-2]) # 뒤에서 네 번째 요소 

print("x1[1:3]:", x1[1:3]) # 1번째 요소부터 3번째 요소까지 슬라이싱
```

```python
# 2차원 배열의 인덱싱 및 슬라이싱 - 동시 사용에 유의하라. 
x2 = np.random.random(size = (10, 5))

print("x2[0, 1]", x2[0, 1])
print("x2[1:4, 2]\n", x2[1:4, 2])
print("x2[0, 2:4]\n", x2[0, 2:4])
print("x2[0:3, 2:4]\n", x2[0:3, 2:4])
```

#### 유니버설 함수: 개별 요소에 반복된 연산을 빠르게 수행하는 것을 주 목적으로 하는 함수이다. 

```python
# 사용 기초 - 크기가 일치됨을 확인하라. 
x = np.array([1, 2, 3, 4])
y = np.array([4, 3, 2, 1])

print("x + y = ", x + y)
print("x - y = ", x - y)
print("x * y = ", x * y)
print("x / y = ", x / y)
print("log x = ", np.log(x))   # 이후 로그 변환할 때 사용하게 된다. 기억합시다. 
```

```python
# 루프와 유니버설 함수의 속도 비교
arr_1 = np.random.randint(1, 10, size = 10 ** 8)
arr_2 = np.random.randint(1, 10, size = 10 ** 8)
import time
### 리스트를 사용하여 벡터의 덧셈 구현
t1 = time.time()
output = []
for val1, val2 in zip(arr_1, arr_2):
    output.append(val1 + val2)
t2 = time.time()
print("반복문을 사용한 경우:", round(t2 - t1, 4))

### 유니버설 함수를 사용하여 벡터의 덧셈 구현
t1 = time.time()
output = arr_1 + arr_2
t2 = time.time()
print("유니버설 함수를 사용한 경우:", round(t2 - t1, 4))
```

```python
# 브로드캐스팅
print("np.arange(3) + 5:",   #np.arange의 경우 단일 숫자를 받았을 때는 0부터 출력하게 된다. 
      np.arange(3) + 5, '\n')

print("np.ones((3, 3)) + np.arange(3)\n",  # 후자가 3x3 모양으로 변한다. 
      np.ones((3, 3)) + np.arange(3), '\n')

print("np.arange(3).reshape(3, 1) + np.arange(3)\n", # 1*3 형태가 3*1 형태로 변환
      np.arange(3).reshape(3, 1) + np.arange(3))
# reshape함수는 행렬의 위치를 뒤바꾼다. 회전의 개념으로 생각하면 쉽다. 
```

```python
# 브로드캐스팅 예시: z - normalization
X = np.random.random((10, 3)) # 0부터 1까지 10*3 모양 array 반환. 
Xmean = X.mean(axis = 0) # 열별 평균 크기: (1, 3)
Xstd = X.std(axis = 0) # 열별 표준편차 크기: (1, 3)

Z = (X - Xmean) / Xstd  # 평균과 표준편차가 각각 10*3모양으로 바뀌게 된다. 

print("X:", X)
print("\nXmean:\n", Xmean)
print("\nXstd:\n", Xstd)
print("\nZ:\n", Z)
```

```python
# 비교 연산자
L = np.array([1, 2, 3, 4, 5])
cond = L >= 3   # 조건을 만들게 됨. 
print(sum(cond)) # 조건을 만족하는 요소의 개수, 객체가 포함 부여된 조건이다. 
cond # cond 자체는 부울 타입의 array이다. 
type(cond)
```
