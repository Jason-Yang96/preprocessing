pandas.get_dummies는 숫자 표현 변수를 인식하지 못하기 때문에 문자열 변수로 바꿔야 한다.
유사도(거리)
1) 유클리디안 거리: 점과의 최단 거리. 
2) 맨하탄 거리: 수직+수평 이동 거리의 합. 리커트 척도의 경우 많이 사용한다. 정수형 데이터...
3) 코사인 유사도: 방향 유사도. 상품 추천의 경우 방향 유사도 측정. 스케일을 고려하지 않음. 각도만 필요하기 때문에 각이 커지면, 유사도가 줄어든다.
4) 매칭유사도: 전체 특징 중 일치하는 비율을 고려함. 
5) 자카드 유사도: 둘 중 하나라도 1을 가지는 특징 중 일치하는 비율. 희소한 이진형 데이터. 텍스트/상품 등등. 페라리를 가진 사람도 가지지 않은 사람은 유사하다. 둘다 없다면 유사하다고 생각하냐? 아니다. 매우 희소한 것을 가지면 유사할 가능성이 높다는 것을 포함하고 있다.  

#### 계층적 군집화
기본개념: 개별 샘플을 군집으로 간주, 거리가 가까운 두 군집을 순차적으로 묶음. 
군집간 거리: 최단 연결법, 샘플간 거리와는 다른 개념
1) 최단 연결법: 이상치에 민감. 계산량이 많다. 모든 거리를 다 계산하기 때문에
2) 최장 연결법: 위와 마찬가지
3) 평균 연결법: 이상치는 둔감. 계산량은 조금 적은 편
4) 중심 연결법: 중심간 거리. 3번의 계산. 
5) 와드 연결법: 
군집 크기를 비슷하게 만드는 효과.
덴드로 그램: 계층 군집화 과정을 트리 형태로 보여준다.  



장단점: 샘플간 거리만 있으면, 군집화가 가능하다. 다양한 거리 척도 활용 가능. 임의성이 존재하지 않다. 많은 계산량. 군집 개수 설정에 대한 제약 존재.

### 고객 특성에 따른 군집화

```python
df = pd.read_csv("Telco_customer_info.csv")
df.head()
```

```python
df.set_index('customerID', inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
df.columns
```

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters = 3,
                                   affinity = 'euclidean',
                                   linkage = 'ward').fit(df) 
```

```python
df['군집정보'] = clusters.labels_
```

```python
df['군집정보'].head()
```

```python
df.groupby(['군집정보'])[['MonthlyCharges', 'TotalCharges']].mean()
```

```python
df.groupby(['군집정보'])[['SeniorCitizen', 'StreamingTV_Yes']].mean()
```

### 구매기록 기준 고객 군집화

```python
df = pd.read_csv("베스트셀러_도서구매기록.txt", sep = "\t", engine = "python")
```

```python
df.head()
```

```python
matrix_df = pd.crosstab(index = df['회원번호'], columns = df['책제목'])
matrix_df.head()
```

```python
# 군집화 모델 인스턴스화 및 학습
from sklearn.cluster import AgglomerativeClustering as AC
clustering_model = AC(n_clusters = 10,
                      affinity = "jaccard",
                      linkage = "average")
clustering_model.fit(matrix_df)
```

```python
cluster_labels = clustering_model.labels_
cluster_labels
```

```python
# 회원별 소속 군집 확인
cluster_info = pd.DataFrame({"회원ID":matrix_df.index, "소속군집":cluster_labels})
cluster_info.head()
```

```python
cluster_info['소속군집'].value_counts() # 대다수가 0번 군집에 속함
```

```python
matrix_df_with_cluster_info = pd.merge(matrix_df, cluster_info, left_index = True, right_on = '회원ID')
matrix_df_with_cluster_info.head()
```

```python
matrix_df_with_cluster_info.groupby(['소속군집'])[matrix_df.columns].mean()
```

```python
# 군집별로 가장 많이 구매하는 책 목록 확인
matrix_df_with_cluster_info.groupby(['소속군집'])[matrix_df.columns].mean().idxmax(axis = 1)
```
